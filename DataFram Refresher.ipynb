## Filter DF Based on a condition

```python
df = pd.read_csv('employees.csv',parse_dates = ['Start Date','Last Login Time'])
df['Senior Management']= df['Senior Management'].astype('bool')
df['Gender'] = df['Gender'].astype('category')
df.head(3)
```

```python
# Select start datae for sr. mgt before 1900-01-01

data1 = df['Senior Management']
data2 = df['Start Date'] < '1990-01-01' # sr. management start date less than/before 1990-01-01

df[data1 & data2]
```

```python
data1 = df['First Name'] == 'Robert' # select first name = robert
data2 = df['Team'] == 'Client Services' # in client service
data3 = df['Start Date'] > '2016-06-01' # start date > 2016-06-01

df[data1 & data2 | data3]
```

## Apply method on single col

```python
def convert_to_string_and_add_millions(number):
    return str(number) + "MILLIONS!"
```

```python
columns = ["Box_Office", "Bond_Actor_Salary"]

for col in columns:
    df[col] = df[col].apply(convert_to_string_and_add_millions)
    
df.head(3)
```

## apply()  method with row values

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor =="Pierce Brosman":
        return "The best"
    
    elif actor == " Roger Moore" and budget > 40:
        return "Enjoyable"
    
    else:
        return "I have no clue"
    
df.apply(good_movie, axis = 'columns')
```
